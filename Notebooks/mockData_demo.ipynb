```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Constants
NUM_WEEKS = 52
NUM_TFs = 10  # Number of time frames
SEAT_CAPACITY = 100
DAYS_IN_WEEK = 7
NUM_DAYS = NUM_WEEKS * DAYS_IN_WEEK

# Set the base demand rate for each TF (we will adjust this further using seasonality and randomness)
base_demand_rates = np.linspace(1, 10, NUM_TFs)

# Seasonality factors for each week of the year (sinusoidal for simplicity)
# Peaks around week 26 (mid-year) and troughs at the start/end of the year
seasonality_factors = (np.sin(np.linspace(0, 2 * np.pi, NUM_WEEKS)) + 1) / 2 + 0.5

# We'll store the demand in a DataFrame for ease of plotting
demand_data = pd.DataFrame(index=pd.date_range(start='2024-01-01', periods=NUM_DAYS), columns=[f'TF{tf}' for tf in range(NUM_TFs)])

# Populate the DataFrame with simulated demand
demand_rate_per_tf = {}
np.random.seed(42)  # For reproducibility
for day in range(NUM_DAYS):
    week_of_year = day // DAYS_IN_WEEK
    seasonality_factor = seasonality_factors[week_of_year]
    sum_demand_rates = 0
    day_demand = {}
    for tf in range(NUM_TFs):
        # Adjust base demand rate with seasonality and some randomness
        demand_rate = base_demand_rates[tf] * seasonality_factor * np.random.uniform(1.8, 2.2)
        sum_demand_rates += demand_rate
        # Generate the demand for this day and TF
        # demand = np.random.poisson(demand_rate)
        # Store in DataFrame
        # demand_data.iloc[day, tf] = min(demand, SEAT_CAPACITY)
        # Store in dictionary
        day_demand[tf] = demand_rate   # Ensure demand does not exceed seat capacity

    demand_rate_per_tf[day] = day_demand
    # Print the total demand rate for debugging
    # print("Total demand rate for day", day, ":", sum_rates)
    

# # Plot the demand for each TF as a function of date
# plt.figure(figsize=(14, 8))
# for tf in range(NUM_TFs):
#     plt.plot(demand_data.index, demand_data[f'TF{tf}'], label=f'TF{tf}')
# plt.xlabel('Date')
# plt.ylabel('Number of Bookings')
# plt.title('Simulated Demand for Flight FL100 Across Different Time Frames')
# plt.legend()
# plt.show()
```
